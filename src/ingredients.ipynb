```python
import json
import pandas as pd
```

```python
df = pd.read_csv('../data/cleaned_cuisines.csv', index_col=0)
df.head()
```

```python
features = [col for col in df.columns if col != 'cuisine']
print(features)
print(len(features))
```

```python
with open('../data/ingredients.json', 'w', encoding='utf-8') as f:
    json.dump(features, f, ensure_ascii=False)
```
